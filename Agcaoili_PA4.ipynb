```python
import pandas as pd
```

```python
boardexams = pd.read_csv('board2.csv')
boardexams
```

```python
Instru = boardexams.loc[(boardexams['Track'] == 'Instrumentation') & (boardexams['Hometown'] == 'Luzon') & (boardexams['Electronics'] > 70), ['Name', 'GEAS', 'Electronics']]

Instru
```

```python
boardexams['Average'] = boardexams[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1)
```

```python
Mindy = boardexams.loc[(boardexams['Gender'] == 'Female') & (boardexams['Average'] >= 55), ['Name', 'Track', 'Electronics', 'Average']]

Mindy
```

```python
import matplotlib.pyplot as plt
```

```python
track_avg = boardexams.groupby('Track')['Average'].mean()

plt.figure(figsize=(8, 5))
plt.bar(track_avg.index, track_avg.values, color='lightgreen')
plt.title('Average Grades by Track')
plt.xlabel('Track')
plt.ylabel('Average Grade')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
gender_avg = boardexams.groupby('Gender')['Average'].mean()

plt.figure(figsize=(6, 4))
plt.bar(gender_avg.index, gender_avg.values, color='salmon')
plt.title('Average Grades by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Grade')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
home_avg = boardexams.groupby('Hometown')['Average'].mean()

plt.figure(figsize=(6, 4))
plt.bar(home_avg.index, home_avg.values, color='skyblue')
plt.title('Average Grades by Hometown')
plt.xlabel('Hometown')
plt.ylabel('Average Grade')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
